```python
df_pos = pd.read_csv(r"C:\Users\hamza\Masaüstü\Mlops_Lecture_Project\data_folder\POS_CASH_balance.csv", encoding="unicode_escape")

pos_agg = df_pos.groupby("SK_ID_CURR")["SK_ID_PREV"].count().reset_index()
pos_agg = pos_agg.rename(columns={"SK_ID_PREV": "POS_COUNT"})

for col, func, name in [
    ("SK_DPD", "max", "SK_DPD_MAX"),
    ("SK_DPD", "mean", "SK_DPD_MEAN"),
    ("SK_DPD_DEF", "max", "SK_DPD_DEF_MAX"),
    ("SK_DPD_DEF", "mean", "SK_DPD_DEF_MEAN"),
    ("CNT_INSTALMENT", "mean", "CNT_INSTALMENT_MEAN"),
    ("CNT_INSTALMENT_FUTURE", "mean", "CNT_INSTALMENT_FUTURE_MEAN"),
]:
    df = df_pos.groupby("SK_ID_CURR")[col].agg(func).reset_index()
    df = df.rename(columns={col: name})
    pos_agg = pos_agg.merge(df, on="SK_ID_CURR", how="left")

dummies = pd.get_dummies(df_pos["NAME_CONTRACT_STATUS"])
dummies["SK_ID_CURR"] = df_pos["SK_ID_CURR"]
status_counts = dummies.groupby("SK_ID_CURR").sum().reset_index()
pos_agg = pos_agg.merge(status_counts, on="SK_ID_CURR", how="left")
```

```python
pos_agg
```

```python
df_inst = pd.read_csv(r"C:\Users\hamza\Masaüstü\Mlops_Lecture_Project\data_folder\installments_payments.csv", encoding="unicode_escape")

inst_agg = df_inst.groupby("SK_ID_CURR")["SK_ID_PREV"].count().reset_index()
inst_agg = inst_agg.rename(columns={"SK_ID_PREV": "INST_COUNT"})

df_inst["DAYS_LATE"] = df_inst["DAYS_ENTRY_PAYMENT"] - df_inst["DAYS_INSTALMENT"]
df_inst["PAYMENT_RATIO"] = df_inst["AMT_PAYMENT"] / df_inst["AMT_INSTALMENT"]

for col, func, name in [
    ("DAYS_LATE", "mean", "DAYS_LATE_MEAN"),
    ("DAYS_LATE", "max", "DAYS_LATE_MAX"),
    ("PAYMENT_RATIO", "mean", "PAYMENT_RATIO_MEAN"),
    ("AMT_INSTALMENT", "mean", "AMT_INSTALMENT_MEAN"),
    ("AMT_PAYMENT", "mean", "AMT_PAYMENT_MEAN"),
]:
    df = df_inst.groupby("SK_ID_CURR")[col].agg(func).reset_index()
    df = df.rename(columns={col: name})
    inst_agg = inst_agg.merge(df, on="SK_ID_CURR", how="left")
```
