```python
from pytorch_tabnet.tab_model import TabNetClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# Continuous cols: scale normally
# Categorical cols: label-encode then pass cat_idxs/cat_dims so TabNet embeds them
cat_cols  = str_cols
cont_cols = [c for c in X_train.columns if c not in cat_cols]

# Scale continuous, encode categoricals for embedding
scaler = StandardScaler()
X_train_cont = scaler.fit_transform(X_train[cont_cols].values.astype(np.float32))
X_test_cont  = scaler.transform(X_test[cont_cols].values.astype(np.float32))

X_train_cat = X_train[cat_cols].values.astype(np.int64)
X_test_cat  = X_test[cat_cols].values.astype(np.int64)

X_train_np = np.hstack([X_train_cont, X_train_cat])
X_test_np  = np.hstack([X_test_cont,  X_test_cat])

y_train_np = y_train.values
y_test_np  = y_test.values

# cat_idxs: positions of categorical columns (after continuous cols)
cat_idxs = list(range(len(cont_cols), len(cont_cols) + len(cat_cols)))
# cat_dims: number of unique values per categorical + 1 (for unknown)
cat_dims = [int(X_train[c].nunique()) + 1 for c in cat_cols]

tabnet = TabNetClassifier(
    n_d=128, n_a=128,
    n_steps=7,
    gamma=1.3,
    cat_idxs=cat_idxs,
    cat_dims=cat_dims,
    cat_emb_dim=4,
    optimizer_params=dict(lr=2e-3),
    scheduler_fn=torch.optim.lr_scheduler.CosineAnnealingLR,
    scheduler_params={"T_max": 200},
    mask_type="entmax",
    device_name=device,
    verbose=20,
)

tabnet.fit(
    X_train_np, y_train_np,
    eval_set=[(X_test_np, y_test_np)],
    eval_metric=["auc"],
    max_epochs=200,
    patience=30,
    batch_size=2048,
    virtual_batch_size=256,
    weights=1,
)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, brier_score_loss, f1_score
from scipy.stats import ks_2samp
import numpy as np

tabnet_proba = tabnet.predict_proba(X_test_np)[:, 1]

# Find threshold that maximises F1 for the minority class
thresholds = np.arange(0.1, 0.6, 0.01)
f1s = [f1_score(y_test_np, (tabnet_proba >= t).astype(int), pos_label=1) for t in thresholds]
best_t = thresholds[np.argmax(f1s)]
print(f"Best threshold: {best_t:.2f}  (default was 0.5)")

tabnet_preds = (tabnet_proba >= best_t).astype(int)

auc   = roc_auc_score(y_test_np, tabnet_proba)
gini  = 2 * auc - 1
ks    = ks_2samp(tabnet_proba[y_test_np == 1], tabnet_proba[y_test_np == 0]).statistic
brier = brier_score_loss(y_test_np, tabnet_proba)

print("=== TabNet ===")
print(f"Accuracy    : {accuracy_score(y_test_np, tabnet_preds):.4f}")
print(f"ROC-AUC     : {auc:.4f}")
print(f"Gini        : {gini:.4f}")
print(f"KS Statistic: {ks:.4f}")
print(f"Brier Score : {brier:.4f}")
print()
print(classification_report(y_test_np, tabnet_preds))
```

```python
import torch
import torch.nn as nn
from pytorch_tabular import TabularModel
from pytorch_tabular.models import FTTransformerConfig
from pytorch_tabular.config import DataConfig, OptimizerConfig, TrainerConfig
from lightning.pytorch.callbacks import Callback
from tqdm.notebook import tqdm as tqdm_notebook

torch.set_float32_matmul_precision("high")

class EpochBar(Callback):
    def on_train_start(self, trainer, pl_module):
        self.bar = tqdm_notebook(total=trainer.max_epochs, desc="FT-Transformer", unit="epoch")
    def on_train_epoch_end(self, trainer, pl_module):
        m = trainer.callback_metrics
        self.bar.set_postfix(
            train_loss=float(m.get("train_loss", 0)),
            val_loss=float(m.get("valid_loss", 0)),
        )
        self.bar.update(1)
    def on_train_end(self, trainer, pl_module):
        self.bar.close()

train_df = X_train.reset_index(drop=True).copy()
test_df  = X_test.reset_index(drop=True).copy()
train_df["TARGET"] = y_train.reset_index(drop=True).astype(int)
test_df["TARGET"]  = y_test.reset_index(drop=True).astype(int)

# Separate true categoricals from continuous
# str_cols were label-encoded — treat them as categoricals so FT-Transformer embeds them properly
cat_cols  = str_cols  # label-encoded columns from earlier cell
cont_cols = [c for c in X_train.columns if c not in cat_cols and c != "TARGET"]

train_df[cont_cols] = train_df[cont_cols].astype("float32")
test_df[cont_cols]  = test_df[cont_cols].astype("float32")
train_df[cat_cols]  = train_df[cat_cols].astype(str)
test_df[cat_cols]   = test_df[cat_cols].astype(str)

# Class weight for imbalance
n_neg = int((train_df["TARGET"] == 0).sum())
n_pos = int((train_df["TARGET"] == 1).sum())
class_weight = torch.tensor([1.0, n_neg / n_pos], device="cuda")
weighted_loss = nn.CrossEntropyLoss(weight=class_weight)
print(f"Categorical cols: {len(cat_cols)}  |  Continuous cols: {len(cont_cols)}")
print(f"Class weight for defaulters: {n_neg/n_pos:.1f}x")

data_config = DataConfig(
    target=["TARGET"],
    continuous_cols=cont_cols,
    categorical_cols=cat_cols,
)

trainer_config = TrainerConfig(
    max_epochs=100,
    batch_size=64,
    accelerator="gpu",
    devices=1,
    precision="bf16-mixed",
    early_stopping="valid_loss",
    early_stopping_patience=10,
)

optimizer_config = OptimizerConfig(
    lr_scheduler="CosineAnnealingLR",
    lr_scheduler_params={"T_max": 60},
)

model_config = FTTransformerConfig(
    task="classification",
    num_attn_blocks=8,
    num_heads=8,
    attn_dropout=0.1,
    ff_dropout=0.1,
    learning_rate=1e-4,
)

ft_transformer = TabularModel(
    data_config=data_config,
    model_config=model_config,
    optimizer_config=optimizer_config,
    trainer_config=trainer_config,
)

ft_transformer.fit(
    train=train_df,
    validation=test_df,
    loss=weighted_loss,
    callbacks=[EpochBar()],
)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, brier_score_loss, f1_score
from scipy.stats import ks_2samp
import numpy as np

ft_result = ft_transformer.predict(test_df)
ft_proba  = ft_result["TARGET_1_probability"].values
ft_true   = test_df["TARGET"].values

# Find best threshold by F1 on minority class
thresholds = np.arange(0.1, 0.7, 0.01)
f1s = [f1_score(ft_true, (ft_proba >= t).astype(int), pos_label=1) for t in thresholds]
best_t = thresholds[np.argmax(f1s)]
print(f"Best threshold: {best_t:.2f}  (default was 0.5)")
ft_preds = (ft_proba >= best_t).astype(int)

auc   = roc_auc_score(ft_true, ft_proba)
gini  = 2 * auc - 1
ks    = ks_2samp(ft_proba[ft_true == 1], ft_proba[ft_true == 0]).statistic
brier = brier_score_loss(ft_true, ft_proba)

print("=== FT-Transformer ===")
print(f"Accuracy    : {accuracy_score(ft_true, ft_preds):.4f}")
print(f"ROC-AUC     : {auc:.4f}")
print(f"Gini        : {gini:.4f}")
print(f"KS Statistic: {ks:.4f}")
print(f"Brier Score : {brier:.4f}")
print()
print(classification_report(ft_true, ft_preds))
```
